```python
!pwd
from PIL import Image
import mxnet as mx
import numpy as np

img = Image.open('../examples/multi-line_cn1.png').convert('L').convert('RGB')
img = np.array(img)
# h, w
h, w, chl = img.shape
assert chl == 3
img = mx.nd.array(img)
# img = img.transpose((2, 0, 1))
# with open('../examples/multi-line_cn1.png', 'rb') as fp:
#     img = fp.read()
# img = mx.image.imdecode(img)
print(f'img shape: {img.shape}')
aug = mx.image.CreateAugmenter(data_shape=(3, h, w), brightness=0.001, contrast=0.001, saturation=0.001,
                              hue=0.05, pca_noise=0.1, inter_method=2)
for a in aug:
    img = a(img)
img = img.astype('uint8').asnumpy()
print(img.shape)
img = Image.fromarray(img).convert('L')
img
```
